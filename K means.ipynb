```python
import pandas as pd
```

```python
data = pd.read_csv('test_data_hidden.csv')
```

```python
from sklearn.cluster import KMeans
```

```python
clu = KMeans(n_clusters=5)
clu.fit(data.iloc[:,0:2])
```

```python
clu.cluster_centers_
```

```python
clu.predict(data.iloc[:,0:2]).tolist()
```

```python
cost = []
for i in range(1,20):
    clu = KMeans(n_clusters=i)
    clu.fit(data.iloc[:,0:2])
    cost.append(clu.inertia_)
```

```python
import matplotlib.pyplot as plt


plt.plot(range(1,20),cost,color='green')
plt.show()
```
