```python
import pandas as pd
```

```python
data = pd.read_csv('Housing.csv')
```

```python
x = pd.DataFrame([[1,'one'],[2,'two'],[3,'Three'],[4,'four']],columns=['num','val'])
```

```python
y = pd.DataFrame([[1,'ten'],[20,'twenty'],[30,'Thirty'],[4,'forty']],columns=['num','val * 10'])
```

```python
# merging
pd.merge(x,y,how='outer',on='num')
```

```python
# grouping data
data.groupby('bathrooms').count()

```

```python
# applying custom functions on data
def fun(x):
    if x < 7000000:
        return "budget house"
    if x > 7000000 and x < 13000000:
        return "mid level budget"
    if x > 13000000:
        return "expensive house"
    else:
        return "Error"
    

    
data['price'].apply(lambda x:fun(x))
```

```python
# pivot tables
data.pivot_table(values='area',index='bathrooms',aggfunc='count')
```
