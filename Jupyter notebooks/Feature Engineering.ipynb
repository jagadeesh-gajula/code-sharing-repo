```python
import pandas as pd
```

```python
data = pd.read_csv('Housing.csv')
```

```python
data = pd.get_dummies(data)
```

```python
target = data.price
inp = data.drop('price',axis=1)
```

```python
import numpy as np
import statsmodels.api as sm
```

```python
inp = sm.add_constant(inp)
```

```python
inp = inp.drop('basement_no',axis=1)
model = sm.OLS(target,inp)
model.fit().summary()
```

```python
inp = inp.drop('mainroad_no',axis=1)
model = sm.OLS(target,inp)
model.fit().summary()
```

```python
inp = inp.drop('guestroom_no',axis=1)
model = sm.OLS(target,inp)
model.fit().summary()
```

```python
inp = inp.drop('furnishingstatus_unfurnished',axis=1)
model = sm.OLS(target,inp)
model.fit().summary()
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression


lin = LinearRegression()

model = RFE(lin,n_features_to_select=7)

model.fit(inp,target)
```

```python
from sklearn.metrics import r2_score

for i in range(1,inp.shape[1]):
    lin = LinearRegression()
    model = RFE(lin,n_features_to_select=i)
    model.fit(inp,target)
    selected_features = [i for i,j in zip(inp.columns,model.support_) if j == True]
    lin.fit(inp[selected_features],target)
    r2 = r2_score(lin.predict(inp[selected_features]),target)
    print(r2)
```
