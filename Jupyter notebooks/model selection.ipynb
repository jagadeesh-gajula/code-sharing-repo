```python
import pandas as pd


data = pd.read_csv('test_data_hidden.csv')
```

```python
data
```

```python
x = data.drop('Class',axis=1)
y = data.Class
```

```python
%%time
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

params = {'criterion':('gini','entropy'),'max_depth':[2,3,4,5],'min_samples_split':[2,4,5],
         'min_samples_leaf':[2,5,6],'random_state':[2,56,90],'max_leaf_nodes':[4,5]}

tree_model = DecisionTreeClassifier()
model = GridSearchCV(tree_model,params,scoring='accuracy',cv=2,verbose=1,n_jobs=8)

model.fit(x,y)

```

```python
model.best_params_
```
