```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('loan.csv')
```

```python
plt.figure(figsize=(15,7))
(data.isnull().sum()/data.shape[0] * 100).plot(kind='bar')
plt.show()
```

```python
cache = data.columns[data.isnull().sum()/data.shape[0] * 100 > 97].tolist()
```

```python
data = data.drop(cache,axis=1)
```

```python
int_cols = [i for i in data.columns if data[i].dtype == 'int64']
float_cols = [i for i in data.columns if data[i].dtype == 'float64']
object_cols = [i for i in data.columns if data[i].dtype == 'object']
```

```python
for i in int_cols:
    print(i)
    print(data[i].describe())
    sns.distplot(data[i])
    plt.yscale('log')
    plt.show()
    
```

```python
for i in float_cols:
    print(i)
    print(data[i].describe())
    plt.hist(data[i])
    plt.yscale('log')
    plt.show()
    
```

```python
data = data[object_cols].fillna('na')
```

```python
for i in object_cols:
    print(i)
    print(data[i].describe())
    data[i].value_counts().plot(kind='bar')
    plt.show()
    
```
