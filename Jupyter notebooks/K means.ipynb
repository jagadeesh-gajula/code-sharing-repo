```python
import pandas as pd
```

```python
data = pd.read_csv('test_data_hidden.csv')
```

```python
data
```

```python
%%time
from sklearn.cluster import KMeans



t = []
for i in range(1,25):
    model = KMeans(n_clusters=i)
    model.fit(data)
    print(model.inertia_)
    t.append(model.inertia_)
```

```python
import matplotlib.pyplot as plt


plt.scatter(range(1,25),t)
```
