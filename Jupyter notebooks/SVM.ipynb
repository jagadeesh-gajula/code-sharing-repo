```python
import pandas as pd

```

```python
data = pd.read_csv('test_data_hidden.csv')
```

```python
data
```

```python
data.Class.value_counts().plot(kind='bar')
```

```python
from sklearn.svm import SVC
```

```python
x = data.drop('Class',axis=1)
```

```python
y = data.Class
```

```python
clf = SVC(C=0.1,kernel='rbf')
```

```python
clf.fit(x,y)
```

```python
pred = clf.predict(x)
```

```python
from sklearn.metrics import f1_score
```

```python
f1_score(y,pred)
```
