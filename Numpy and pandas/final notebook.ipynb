# Data cleaning

```python
dataset = pd.read_csv('loan.csv')
```

```python
dataset.shape
```

```python
# basic cleaning functions
dataset.notnull().sum()
dataset.isnull().sum(axis=0)
dataset.notnull().any()
```

## Filling missing values

```python
dataset.fillna(100).head()
```

```python
dataset.isnull().any()
```

## deleting columns with 100% missing values

```python
# dropping 100% missing values columns
missing = dataset.columns[(dataset.isnull().sum() / dataset.shape[0])* 100 == 100].tolist()
```

```python
dataset = dataset.drop(missing,axis=1)
```

```python
# dropping 95% missing values in columns
missing = dataset.columns[(dataset.isnull().sum() / dataset.shape[0])* 100 > 95].tolist()
```

```python
dataset = dataset.drop(missing,axis=1)
```

```python
dataset.shape
```

```python
# dropping columns with only 1 unique values
unique_cols = dataset.columns[dataset.nunique()/dataset.shape[0] * 100 == 100]
```

```python
dataset = dataset.drop(unique_cols,axis=1)
```

```python
dataset.shape
```

```python
single_unique = dataset.columns[dataset.nunique() == 1].tolist()
```

```python
dataset = dataset.drop(single_unique,axis=1)
```
